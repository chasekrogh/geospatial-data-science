# Below is a choropleth map of Multnomah county based on pop_density

```python
# Download data
mult_pop = products.ACS(2019).from_county('Multnomah County, OR', level='tract',
                                        variables=['B01003_001E']) # don't worry about the deprecation message!
```

```python
mult_pop['pop_density'] = 1e6 * mult_pop['B01003_001E'] / mult_pop['geometry'].area

from mpl_toolkits.axes_grid1 import make_axes_locatable

# Plot map
f, ax = plt.subplots(1, 1, figsize=(10,10))

# These two lines make the colorbar the same size as the axes.
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

mult_pop.plot(column='pop_density', ax=ax, cmap='viridis', legend=True, cax=cax, scheme = "quantiles", k = 8)
```

The map above is a choropleth scheme base on population density by tract level in Multnomah County, Oregon. As you can see, most of the population is concentrated near downtown Portland around the Willamette river.
